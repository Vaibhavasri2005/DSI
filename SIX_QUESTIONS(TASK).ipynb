Q.no:2(1):
Using various functions in numpy library, mathematically calculate the values for a
normal distribution and create Histograms to plot the probability distribution curve


```python

import numpy as np
import matplotlib.pyplot as plt

# Generate random data for a normal distribution
mean = 0
std_dev = 1
num_samples = 1000

data = np.random.normal(mean, std_dev, num_samples)

# Plotting the histogram
plt.hist(data, bins=30, density=True, alpha=0.7, color='blue')

# Calculating the theoretical probability distribution curve
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = np.exp(-0.5 * ((x - mean) / std_dev) ** 2) / (std_dev * np.sqrt(2 * np.pi))

# Plotting the probability distribution curve
plt.plot(x, p, 'k', linewidth=2)

plt.title('Normal Distribution and Probability Distribution Curve')
plt.xlabel('Values')
plt.ylabel('Probability Density')
plt.show()
```

Q.no:2(2):using plt.contour(), plt.contourf(), plt.imshow(), plt.colorbar(), plt.clabel() funtions visualize a contour plot

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a synthetic 2D dataset
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = np.sin(np.sqrt(X**2 + Y**2))

# Create a contour plot using plt.contour()
plt.figure(figsize=(10, 5))
plt.subplot(231)
contour = plt.contour(X, Y, Z, cmap='viridis')
plt.title('plt.contour()')

# Create a filled contour plot using plt.contourf()
plt.subplot(232)
contourf = plt.contourf(X, Y, Z, cmap='viridis')
plt.title('plt.contourf()')

# Create an image plot using plt.imshow()
plt.subplot(233)
imshow = plt.imshow(Z, extent=[-5, 5, -5, 5], origin='lower', cmap='viridis')
plt.title('plt.imshow()')

# Add a colorbar to the image plot
plt.colorbar(imshow, ax=plt.gca(), orientation='vertical')

# Create labeled contours using plt.clabel()
plt.subplot(234)
contour_labeled = plt.contour(X, Y, Z, cmap='viridis')
plt.clabel(contour_labeled, inline=True, fontsize=8)
plt.title('plt.clabel()')

# Create a combination plot with contour and filled contour using plt.contour() and plt.contourf()
plt.subplot(235)
contour_combined = plt.contour(X, Y, Z, colors='black', linewidths=0.5)
contourf_combined = plt.contourf(X, Y, Z, cmap='viridis')
plt.colorbar(contourf_combined, ax=plt.gca(), orientation='vertical')
plt.title('Combined plt.contour() and plt.contourf()')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
# Create a Basemap instance for the world map
m = Basemap(projection='mill', llcrnrlat=-60, urcrnrlat=90, llcrnrlon=-180, urcrnrlon=180, resolution='c')
```

```python
# Example data (longitude, latitude)
lon = [0, 30, -45, 60]
lat = [10, -20, 40, 80]

# Plot points on the map
x, y = m(lon, lat)
m.scatter(x, y, marker='o', color='r', label='Sample Points')

# Draw coastlines and country borders
m.drawcoastlines()
m.drawcountries()

# Show the legend
plt.legend()

# Show the plot
plt.show()
```

```python
# Add grid lines
m.drawparallels(range(-90, 91, 30), labels=[1,0,0,0])
m.drawmeridians(range(-180, 181, 60), labels=[0,0,0,1])

# Add a title
plt.title('Sample Map with Basemap')

# Show the plot
plt.show()
```
